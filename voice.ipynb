```python
from gtts import gTTS
import IPython.display as ipd
import random

def get_irrigation_alert(soil_moisture):
    if soil_moisture < 40:
        return "The soil is too dry. Start irrigation."
    elif 40 <= soil_moisture < 70:
        return "The soil moisture is at a moderate level. Irrigation may be needed."
    else:
        return "Soil moisture is sufficient. No irrigation needed."

# Simulating random soil moisture levels (between 20% and 100%)
soil_moisture = random.randint(20, 100)
alert_message = get_irrigation_alert(soil_moisture)

print(f"Soil moisture: {soil_moisture}%")
print(f"Irrigation Alert: {alert_message}")

# Convert the alert to voice
tts = gTTS(text=f"Soil moisture: {soil_moisture} percent. {alert_message}", lang='en')
tts.save("irrigation_alert.mp3")
ipd.display(ipd.Audio("irrigation_alert.mp3"))
```

```python
from gtts import gTTS
import IPython.display as ipd
import random

def get_pest_prediction(crop, humidity, temperature):
    pest_risk = {
        "Rice": {"high_humidity": "High risk of Brown Plant Hopper.", "high_temperature": "Moderate risk of bacterial leaf blight."},
        "Wheat": {"high_humidity": "High risk of rust disease.", "high_temperature": "Moderate risk of aphid infestation."},
        "Cotton": {"high_humidity": "High risk of bollworm attack.", "high_temperature": "Moderate risk of whitefly infestation."}
    }

    alert_messages = []

    if humidity > 80:
        alert_messages.append(pest_risk.get(crop, {}).get("high_humidity", "No high humidity-related risks detected."))
    if temperature > 30:
        alert_messages.append(pest_risk.get(crop, {}).get("high_temperature", "No high-temperature-related risks detected."))

    if not alert_messages:
        return "No pest prediction available for current conditions."

    return " ".join(alert_messages)

# Simulating random humidity (50% to 100%) and temperature (20°C to 40°C)
crop = input("Enter crop name (e.g., Rice): ")
humidity = random.randint(50, 100)
temperature = random.randint(20, 40)

prediction_message = get_pest_prediction(crop, humidity, temperature)

print(f"Humidity: {humidity}% | Temperature: {temperature}°C")
print(f"Pest Prediction for {crop}: {prediction_message}")

# Convert pest alert to speech
tts = gTTS(text=f"Pest prediction for {crop}. Humidity: {humidity} percent, Temperature: {temperature} degrees Celsius. {prediction_message}", lang='en')
tts.save("pest_prediction.mp3")
ipd.display(ipd.Audio("pest_prediction.mp3"))
```

```python
from gtts import gTTS
import IPython.display as ipd
import random

def get_pest_prediction(crop, humidity, temperature):
    pest_risk = {
        "चावल": {"high_humidity": "भूरा पौधा होपर का उच्च जोखिम।", "high_temperature": "जीवाणु पत्ती झुलसा का मध्यम जोखिम।"},
        "गेहूं": {"high_humidity": "रस्ट रोग का उच्च जोखिम।", "high_temperature": "एफिड संक्रमण का मध्यम जोखिम।"},
        "कपास": {"high_humidity": "बोलवर्म हमले का उच्च जोखिम।", "high_temperature": "व्हाइटफ्लाई संक्रमण का मध्यम जोखिम।"}
    }

    alert_messages = []

    if humidity > 80:
        alert_messages.append(pest_risk.get(crop, {}).get("high_humidity", "कोई उच्च आर्द्रता-संबंधी जोखिम नहीं मिला।"))
    if temperature > 30:
        alert_messages.append(pest_risk.get(crop, {}).get("high_temperature", "कोई उच्च तापमान-संबंधी जोखिम नहीं मिला।"))

    if not alert_messages:
        return "वर्तमान परिस्थितियों के लिए कोई कीट भविष्यवाणी उपलब्ध नहीं।"

    return " ".join(alert_messages)

# यादृच्छिक आर्द्रता (50% से 100%) और तापमान (20°C से 40°C) उत्पन्न करना
crop = input("फसल का नाम दर्ज करें (उदा. चावल): ")
humidity = random.randint(50, 100)
temperature = random.randint(20, 40)

prediction_message = get_pest_prediction(crop, humidity, temperature)

print(f"आर्द्रता: {humidity}% | तापमान: {temperature}°C")
print(f"{crop} के लिए कीट भविष्यवाणी: {prediction_message}")

# हिंदी में वॉयस अलर्ट
tts = gTTS(text=f"{crop} के लिए कीट भविष्यवाणी। आर्द्रता: {humidity} प्रतिशत, तापमान: {temperature} डिग्री सेल्सियस। {prediction_message}", lang='hi')
tts.save("pest_prediction_hindi.mp3")
ipd.display(ipd.Audio("pest_prediction_hindi.mp3"))
```
